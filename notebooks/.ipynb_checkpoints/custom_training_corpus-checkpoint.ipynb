```python
import pandas as pd
import numpy as np
import pickle

import random
import string 
import re
from nltk.tokenize import word_tokenize

import spacy
nlp = spacy.load('en')

%matplotlib inline
from matplotlib import pyplot as plt
```

## Preprocess the CEFR word-list

```python
efllex_df = pd.read_csv("./data/misc/EFLLex_lexicon", sep='\t', encoding='utf-8')

# keep only the relevant columns
efllex_df = efllex_df.iloc[:, 0:7]
efllex_headers_kept = ['word','tag','A1','A2','B1','B2','C1']
efllex_df.columns = efllex_headers_kept

# create a new dataframe, only with the CEFR, in order to assign as a label the first non zero occurence
efllex_only_cefr = efllex_df.drop(columns=["word","tag"])
efllex_df['cefr_level'] = efllex_only_cefr.keys()[np.argmax(efllex_only_cefr.values!=0,axis=1)].tolist()
efllex_df = efllex_df.drop(columns=["A1","A2","B1","B2","C1"])

# remove NULL Values from data if any
efllex_df = efllex_df[pd.notnull(efllex_df['word'])]

# further preprocessing to clean the cells
efllex_df['word'] = efllex_df['word'].apply(lambda s : s.replace("-",""))
efllex_df['word'] = efllex_df['word'].apply(lambda s : s.replace("_"," "))
efllex_df['word'] = efllex_df['word'].apply(lambda s : s.strip())
efllex_df['word'] = efllex_df['word'].apply(lambda s : s.lower())

efllex_df = efllex_df.rename(columns={"tag": "pos"})

# rename the POS tags in the UD way
# keep only Noun, Adjective, Adverbs, Verbs because we will examine vocabulary level according to them
# pronouns were not taken under consideration, mainly because there were not many of them in our word-list
for i, row in efllex_df.iterrows():
    if row['pos'] == 'NN':
        row['pos'] = 'NOUN'
    elif row['pos'] == 'JJ':
        row['pos'] = 'ADJ'
    elif row['pos'] == 'VB':
        row['pos'] = 'VERB'
    elif row['pos'] == 'RB':
        row['pos'] = 'ADV'
    elif row['pos'] == 'EX':
        row['pos'] = 'ADV'
    elif row['pos'] == 'WRB':
        row['pos'] = 'ADV'
    else:
        row['pos'] = np.nan
```

```python
# we keep aside MTW for more convenient processing
def get_multi_tokens(word):
    if len(word.split(" "))>1:
        return word
    else:
        return ''
```

```python
# clean the multitokens
def clean_multi_tokens(word):
    if len(word.split(" "))>1:
        return ''
    else:
        return word
```

```python
# get all the multitokens in a seperate list, for seperate processing later
multi_df = efllex_df.copy()
multi_df['word'] = multi_df['word'].apply(lambda s : get_multi_tokens(s))
multi_df['word'].replace('', np.nan, inplace=True)
multi_df.dropna(subset=['word'], inplace=True)

# remove NULL Values from data if any
multi_df = multi_df[pd.notnull(multi_df['pos'])]

# cluster together the categories
# instead of A,B,C we will use numbering, because it will help later to get the max category of a sentence
for i,  row in multi_df.iterrows():
    if row['cefr_level'] == 'A1':
        row['cefr_level'] = 1
    elif row['cefr_level'] == 'A2':
        row['cefr_level'] = 1
    elif row['cefr_level'] == 'B1':
        row['cefr_level'] = 2
    elif row['cefr_level'] == 'B2':
        row['cefr_level'] = 2
    elif row['cefr_level'] == 'C1':
        row['cefr_level'] = 3
    elif row['cefr_level'] == 'C2':
        row['cefr_level'] = 3
        
multi_df.reset_index(drop=True)
```

```python
# keep only the single-word tokens in this part
# null all the multitokens
efllex_df['word'] = efllex_df['word'].apply(lambda s : clean_multi_tokens(s))
efllex_df['word'].replace('', np.nan, inplace=True)
efllex_df.dropna(subset=['word'], inplace=True)

# remove NULL Values from data if any
efllex_df = efllex_df[pd.notnull(efllex_df['pos'])]

# drop rows which have the same word and POS (different pos may indicate different sense) and keep latest entry
efllex_df = efllex_df.drop_duplicates(subset = ['word', 'pos'], keep = 'last').reset_index(drop = True)

# cluster together the categories
# instead of A,B,C we will use numbering, because it will help later to get the max category of a sentence
for i,  row in efllex_df.iterrows():
    if row['cefr_level'] == 'A1':
        row['cefr_level'] = 1
    elif row['cefr_level'] == 'A2':
        row['cefr_level'] = 1
    elif row['cefr_level'] == 'B1':
        row['cefr_level'] = 2
    elif row['cefr_level'] == 'B2':
        row['cefr_level'] = 2
    elif row['cefr_level'] == 'C1':
        row['cefr_level'] = 3
    elif row['cefr_level'] == 'C2':
        row['cefr_level'] = 3
        
efllex_df.reset_index(drop=True)
```

```python
count_df = efllex_df[['pos','word']].groupby('pos').aggregate({'word':'count'}).reset_index().sort_values('word',ascending=False)
count_df.head()
```

```python
# obtain the lemma of each word - according to which we will base our single-word labeling
def lemmatize_text(text):
    doc = nlp(text)
    for token in doc:
        return (token.lemma_)
```

```python
# create a list with all the lemmas
efllex_df['lemma'] = efllex_df.word.apply(lemmatize_text)
```

```python
# preprocessing function
def clean_text(x):
    pattern = r'[^a-zA-z0-9\s]'
    text = re.sub(pattern, '', x)
    return x
```

```python
# clean lemma values
efllex_df['lemma'].str.lower()
efllex_df['lemma'] = efllex_df.lemma.apply(clean_text)
efllex_df = efllex_df[~efllex_df.lemma.str.contains("'")]

# delete entries that contain a number (irrelevant for our cause)
efllex_df =  efllex_df[~efllex_df.lemma.str.contains(r'[0-9]')]
```

## Labelize each sentence according to the CEFR word-list

```python
multi_df = pd.read_pickle('./data/misc/multi_cefr_word_list.pkl')

# keep seperately the multitokens and their labels
multitokens_words = multi_df.word.tolist()
multitokens_cefr = multi_df.cefr_level.tolist()
```

```python
efllex_df = pd.read_pickle('./data/misc/cefr_word_list_lemmatized.pkl')
```

```python
def get_cefr_level(sentence, accepted_tags):

    # in case a multi-token of our lexicon exist in a sentence
    if any(multiword in sentence for multiword in multitokens_words):

        # replace it with its corresponding CEFR level. In that way we also take under consideration the MWTs
        multi_index = [multi_index for multi_index, multiword in enumerate(multitokens_words) if multiword in sentence]

        # we replace the multi-token with its CEFR label and also apply some diacritics in order to identify it from irrelevant numbers in a sentence 
        sentence = sentence.replace(multitokens_words[multi_index[0]], str(multitokens_cefr[multi_index[0]]))

    cefr_labels = []  
    doc = nlp(sentence)

    for token in doc:

        # in case of a multi-token, we shall append its label
        if token.text == '1' or token.text == '2' or token.text == '3':
            cefr_labels.append(int(token.text))

        # many of our instances were labelled as proper nouns by spacy, while our word-list tagged them as nouns
        if token.pos_ == 'PROPN':
            pos_tag = 'NOUN'
        else:
            pos_tag = token.pos_

        # check only for the accepted pos
        if pos_tag in accepted_tags:

            # if the lemma of that token does not exist in my lemma list, skip that sentence
            if not (efllex_df['lemma'].str.contains(token.lemma_.lower()).any()):
                return 0

            else:
                try:
                    # get the index of that word, in my word-list (there maybe more than one indexes, indicating different POS and senses)
                    labeling_indexes = efllex_df[efllex_df['lemma']==token.lemma_.lower()].index.values.tolist()

                    # for more than one indexes, find the index of the word that has the same POS with the token under examination
                    if (len(labeling_indexes)>1):
                        for index_label in labeling_indexes:
                            if efllex_df['pos'][index_label] == pos_tag:
                                cefr_labels.append(int(efllex_df['cefr_level'][index_label]))

                    # if only one index exist, retrieve the cefr label of that one
                    else:
                        cefr_labels.append(int(efllex_df['cefr_level'][labeling_indexes]))

                # in case of any possible error, skip the sentence
                except:
                    return 0

    # return the maximum label value, as the label of the sentence
    if max(cefr_labels) == 3:
        return 'C'
    elif max(cefr_labels) == 2:
        return 'B'
    elif max(cefr_labels) == 1:
        return 'A'
```

# Tatoeba data

```python
# the tatoeba list containing information about the language level of the translators
users_df = pd.read_csv("./data/misc/tatoeba_translators_level.csv", sep='\t', names=['language','level', 'username'], index_col=False, encoding='utf-8') 

# restrict the data to english translations
users_df = users_df.loc[users_df['language'] == 'eng']

# restrict the data to english level of 5 (the highest)
users_df = users_df.loc[(users_df['level'] == '5')]

allowed_users = users_df['username'].tolist()
```

```python
# a dataframe containing the english tatoeba sentences, along with their translator
sentences_df = pd.read_csv("./data/misc/tatoeba_corpus.tsv", sep='\t', names=['id','language', 'sentence','username', 'date1', 'date2'], index_col=False, encoding='utf-8') 

# keep only the translators we retrieved before (based on level)
sentences_df = sentences_df.loc[sentences_df['username'].isin(allowed_users)]
```

```python
sentences_df
```

```python
# drop duplicates
tatoeba = sentences_df.drop_duplicates(subset='sentence', keep="last")
tatoeba = tatoeba["sentence"].tolist()
```

```python
# visualize the lengths of our sentences.
plt.style.use('seaborn-whitegrid')

lengths = [len(line) for line in tatoeba if len(line) > 1]
plt.hist(lengths, bins=np.arange(100), histtype='step', linewidth=1);
```

```python
# shuffle the list because they are sorted on augmenting length
random.shuffle(tatoeba)
random.shuffle(tatoeba)
random.shuffle(tatoeba)

# in that way I will be able to restore from the exact step in a case of broken loop
with open('./data/misc/tatoeba_shuffled.pkl', 'wb') as f:
    pickle.dump(tatoeba, f)
```

```python
# check only for these pos tags
accepted_tags = ['ADJ','ADV','NOUN','VERB']

sentences_df = []
labels_df = []

labels_a = []
labels_b = []
labels_c = []

flag = True
while (flag):
    for step, sent in enumerate(tatoeba):

        # the collection of labeled sentences will happen until for all three labels, 70.000 instances are gathered
        if (len(labels_a)<70000 or len(labels_b)<70000 or len(labels_c)<70000):
            try:
                cefr_label = get_cefr_level(sent, accepted_tags)
            except:
                pass

            if cefr_label:
                sentences_df.append(sent)
                labels_df.append(cefr_label)

                if cefr_label == 'A':
                    labels_a.append(cefr_label)
                elif cefr_label == 'B':
                    labels_b.append(cefr_label)
                elif cefr_label == 'C':
                    labels_c.append(cefr_label)

                # save the lists every 1000 iterations 
                if step % 1000 == 0:
                    with open('./data/misc/tatoeba_generation_checkpoints/sentences_checkpoint.pkl', 'wb') as f:
                        pickle.dump(sentences_df, f)
                    with open('./data/misc/tatoeba_generation_checkpoints/labels_checkpoint.pkl', 'wb') as f:
                        pickle.dump(labels_df, f)

                #statistics 
                print(len(labels_a))
                print(len(labels_b))
                print(len(labels_c))
                print()

        else:
            flag = False
```

```python
# read the final checkpoints
with open('./data/misc/tatoeba_generation_checkpoints/sentences_checkpoint.pkl', 'rb') as f:
    sentences_checkpoint = pickle.load(f)

with open('./data/misc/tatoeba_generation_checkpoints/labels_checkpoint.pkl', 'rb') as f:
    labels_checkpoint = pickle.load(f)
```

```python
tatoeba_data = {
        'Sentences':  sentences_checkpoint,
        'Labels': labels_checkpoint
        }

tatoeba_df = pd.DataFrame (tatoeba_data, columns = ['Sentences','Labels'])
```
